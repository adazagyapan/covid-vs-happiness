### Import the Modules

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

### Import the COVID-19 dataset


```python
corona_dataset_csv = pd.read_csv("/Users/ada/Desktop/Programming/python/personal/Covid Spread vs. Happiness/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

#### Checking the shape of the data frame

```python
corona_dataset_csv.shape
```

```python
### Delete the useless column
```

corona_dataset_csv.drop(["Lat", "Long"], axis=1, inplace=True)

```python
corona_dataset_csv.head(10)
```

### Aggregating the dataset

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.shape
```

### Visualizing data related to a country for example China
Visualization always helps for better understanding of our data.

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

### Calculating a good measure 
Find the spread rate in each country.

```python
corona_dataset_aggregated.loc["China"].plot()
```

#### Plotting only the first 3 days 

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

### Calculating the first derivative of the curve

```python
corona_dataset_aggregated.loc["China"].diff().plot() #.diff() -> the derivative method
```

### Maximum infection rate.

```python
corona_dataset_aggregated.loc["China"].diff().max() #in 24 hours, 16136 new cases have been recorded
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

### Maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head(10)
```

### New dataframe with the only needed column.

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head(10)
```

### Importing the happiness report

### Dropping the useless columns
